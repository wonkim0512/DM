#### 1-1. Review

```python
import pandas as pd
```

```python
# 변수별 설명 : CodeList
pd.read_csv('CodeList-classification.csv').iloc[:, :4]
```

```python
# df
df = pd.read_csv('german_classification.csv').set_index('OBS#')
```

##### Preprocessing : Dummy Variables & Scaling & Change Labels

```python
# dummy varaibles : preprocessed_df
get_dummies_cols = ['CHK_ACCT', 'HISTORY', 'SAV_ACCT', 'EMPLOYMENT', 'JOB']
dummied_df = pd.get_dummies(df, columns=get_dummies_cols)

# scaling
scaling_cols = ['DURATION', 'INSTALL_RATE', 'AGE', 'NUM_CREDITS', 'AMOUNT']
preprocessed_df = dummied_df.copy()
preprocessed_df[scaling_cols] = (preprocessed_df[scaling_cols]-preprocessed_df[scaling_cols].min()) / (preprocessed_df[scaling_cols].max()-preprocessed_df[scaling_cols].min())

# change Label
preprocessed_df.RESPONSE = df.RESPONSE-1

# split X, y 
X, y = preprocessed_df.drop('RESPONSE', axis=1), preprocessed_df.RESPONSE
```

##### Train-Test Split : `sklearn.model_selection.train_test_split`

```python
# import packages
from sklearn.model_selection import train_test_split

# train_test_split with 'stratify'
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, stratify = y, random_state = 1)
```

random_state와 CV의 차이는?
> random_state에서는 모든 데이터가 반영되지 않을 수 있다. 하지만 CV는 등분해서 사용하므로 모든 데이터가 사용된다.

##### `sklearn.linear_model.LogisiticRegression`

```python
from sklearn.linear_model import LogisticRegression

# parameters
class_weight = 'balanced'

# model
model = LogisticRegression(class_weight = class_weight)

# fit
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix

def result(fitted_model, X_test, y_test):
    # pred
    y_pred = fitted_model.predict(X_test)
    
    # score
    score = fitted_model.score(X = X_test, y = y_test)
    
    # conf_mat
    conf_mat = confusion_matrix(y_true = y_test, y_pred = y_pred)
    
    return [y_pred, score, conf_mat]
```

### result
raw_result = result(raw_model, X_test, y_test)

```python
raw_result = result(model, X_test, y_test)
raw_result
```

#### 1-2. RFE : Recursive Feature Elimination `sklearn.feature_selection.RFE`

```python
# import packages
from sklearn.feature_selection import RFE
```

```python
# estimator & selector
selector = RFE(model) # 얘가 알아서 변수를 선택해줌.
```

```python
# fit to selector
selector.fit(X_train, y_train)
# 왜 X_train을 그대로 넣지? 변수 30개 그대로 쓴다고? Yes! 왜냐햐면
# selector는 변수를 추려낸 새로운 모델을 만드는 것이 아니라 변수를 몇개 쓰면 좋을지, 어떤걸 쓰면 좋을지에 대한 report의 개념이라고 보면 된다!
```

```python
# n_features_, ranking, support
print(selector.n_features_)
print(selector.ranking_) # 숫자가 큰 녀석들이 먼저 변수에서 빠지는 것
print(selector.support_) # 1인 애들만, 즉 선택된 변수들만 True로 반환
selector.estimator_
```

```python
# change X to selected_X_train, selected_X_test
selected_cols = selector.support_
X_RFE_test = X_test.loc[:, selected_cols]
```

```python
# fitting new model
# RFE_model
RFE_model = selector.estimator_

# get result
result(RFE_model, X_RFE_test, y_test) # 변수를 24개만 쓰고도 69%를 얻을 수 있다.
```

#### 1-3. PCA : `sklearn.decompostion.pca`

```python
# import packages
from sklearn.decomposition import pca
```

```python
# parameter : n_components
n_components = [5, 10, 15, 20, 25]
```

```python
# pca_model
pca_model = LogisticRegression(class_weight= class_weight)

for n in n_components:
    # decomposer
    decomposer = pca.PCA(n_components = n)
    
    # fit
    decomposer.fit(X_train) # 나의 관심사는 y가 아니라 x의 차원을 줄이는 것이므로 X_train에 대해서만 fit한다.
    
    # .transform
    X_pca_train = decomposer.transform(X_train)
    print(X_pca_train)
    print(n,'principal components are selected: ', decomposer.explained_variance_ratio_, "->", \
          sum(decomposer.explained_variance_ratio_)) # 주성분 하나가 설명하는 비율
    
    
    # fit new estimator
    pca_model.fit(X = X_pca_train, y = y_train)
    
    # result()
    #pca_result = result(pca_model, X_test, y_test) # X_test also needs pca
    X_pca_test = decomposer.transform(X_test)
    pca_result = result(pca_model, X_pca_test, y_test)
    print(pca_result)
    print("*"*100)
```

#### 1-4. 실습 : 변수 선택을 활용한 신용 등급 분류 모델 만들기
---
1. 변수 선택을 적용한 신용 등급 분류 모델 만들기
    * RFE, PCA 중 반드시 하나를 사용할 것
    * 변수 선택 하지 않은 모델(모든 변수를 사용한 모델)과 성능을 비교할 것
    
2. 변수 선택을 활용했을 때의 장점은 무엇인가?
> 적은 변수를 쓰고도 recall score가 충분히 높음. 오히려 더 높아졌음. 또한, 모델의 복잡도가 줄어드므로 새로운 데이터가 들어왔을 때 분류를 잘 할것! 예를 들어, 2차원과 10차원 다항식이 있을 때 10차원은 과적합돼있어서 새로운 데이터가 들어왔을 때 설명력이 낮아질 것

변수선택은 실제로 변수를 줄이는 것이고, PCA는 선택하는 것이라 아니라 조합으로 통해 가장 좋은 축을 찾아내는 것. 즉 PCA는 모든 변수를 다 사용하는 것!

```python
from sklearn.metrics import recall_score

def recall(fitted_model, X_test, y_test):
    # pred
    y_pred = fitted_model.predict(X_test)
    
    # score
    recall = recall_score(y_true = y_test, y_pred = y_pred)
    
    # conf_mat
    conf_mat = confusion_matrix(y_true = y_test, y_pred = y_pred)
    
    #return [y_pred, recall, conf_mat]
    return [recall, conf_mat]
```

```python
from sklearn.linear_model import LogisticRegression

# parameters
class_weight = 'balanced'

# model
model = LogisticRegression(class_weight = class_weight)

# fit
model.fit(X_train, y_train)
```

```python
recall(model, X_test, y_test)
```

```python
# pca_model
pca_model = LogisticRegression(class_weight= class_weight)

for n in n_components:
    # decomposer
    decomposer = pca.PCA(n_components = n)
    
    # fit
    decomposer.fit(X_train) # 나의 관심사는 y가 아니라 x의 차원을 줄이는 것이므로 X_train에 대해서만 fit한다.
    
    # .transform
    X_pca_train = decomposer.transform(X_train)
    print(X_pca_train)
    print(n,'principal components are selected: ', decomposer.explained_variance_ratio_, "->", \
          sum(decomposer.explained_variance_ratio_)) # 주성분 하나가 설명하는 비율
    
    
    # fit new estimator
    pca_model.fit(X = X_pca_train, y = y_train)
    
    # result()
    #pca_result = result(pca_model, X_test, y_test) # X_test also needs pca
    X_pca_test = decomposer.transform(X_test)
    pca_result = recall(pca_model, X_pca_test, y_test)
    print(pca_result)
    print("*"*100)
```

##### Bag of words : `sklearn.feature_extraction.text.CountVectorizer`

##### Word Cloud : `wordcloud.WordCloud`

```python
# import packages
from wordcloud import WordCloud
```

```python
# initialize
wordcloud = WordCloud()
```

```python
# sort bows by count
# vocabs
vocabs = vectorizer.get_feature_names()

# counts
counts = bows.tolist()[0]

# sorted_bows
sorted_bows = sorted(zip(vocabs, counts), key = lambda t: t[1], reverse = True) # count를 기준으로 sort
sorted_bows
```

```python
new_stopwords = ["cobb", "arthur", "ariadn", "fischer", "saito", "yusuf"]
new_sorted_bows = [bow for bow in sorted_bows if bow[0] not in new_stopwords]
new_sorted_bows
```
