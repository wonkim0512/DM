# Hierarchical Clustering 
## 1. sklearn

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
n_cluster = 5
sklearn_hc_model = AgglomerativeClustering(n_clusters = n_cluster)
sklearn_hc_model.fit(X_df)
```

```python
sklearn_hc_model.labels_
```

```python
result_df = X_df.copy()
result_df['sklearn_hc_labels'] = sklearn_hc_model.fit_predict(X_df)
```

## 2. scipy

```python
from scipy.cluster.hierarchy import linkage
```

```python
z = linkage(X_df, 'ward')
z[:10]
```

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram
```

```python
def plot_dendrogram(z, max_d = None, p = None):
    plt.figuresize = (25, 10)
    plt.title("Dendrogram")
    plt.xlabel("Index")
    plt.ylabel("Distance")
    
    if p:
        dendrogram(
            z,
            leaf_rotation = 90.,
            leaf_font_size = 8.,
            truncate_mode = 'lastp',
            p = p,
            show_contracted = True)
        
    else:
        dendrogram(
            z,
            leaf_rotation = 90.,
            leaf_font_size = 8.)
    
    if max_d:
        plt.axhline(max_d, c = "k")
    plt.show()
```

```python
plot_dendrogram(z)
```

```python
plot_dendrogram(z, p =5, max_d = 58000)
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
scipy_hc_labels = fcluster(z, 58000, criterion = 'distance')
set(scipy_hc_labels)
```

```python
result_df['scipy_hc_labels'] = scipy_hc_labels
```
