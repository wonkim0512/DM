#### 1.1. Data Structure
---
##### pd.Series 
1. Labeling이 가능한 1차원 Array Object 
2. 모두 같은 데이터형을 가집니다.

```python
# index가 ['a', 'b', 'c', 'd']이고, 값이 [3, -5, 7, 4]인 Series
s = pd.Series([3, -5, 7, 4], index=['a', 'b', 'c', 'd'])
```

##### pd.DataFrame 
1. Labeling이 가능한 2차원 Array Object
2. column 별로 서로 다른 데이터형을 가질 수 있습니다.

```python
data = {'Country': ['Belgium', 'India', 'Brazil'],
        'Capital': ['Brussels', 'New Delhi', 'Brasília'],
        'Population': [11190846, 1303171035, 207847528]}
```

##### pd.merge
Durable과 Dm_Sheet의 DataFrame을 통합
* 합치는 키가 되는 column을 잘 살필 것
* 합치는 방법을 잘 결정할 것

```python
merged_df = df.merge(durable_df, how="inner", left_on='Member id', right_on='MEM', suffixes=('_', '_durable'))
```

```python
merged_df
```

##### 결측치 처리  - pd.fillna OR delete rows
* NaN이 있는 행부터 찾을 것
* 해당 행의 성질에 따라 행의 값을 채우거나 지울 것
* 위의 과정을 시행할 때 `nan_filled_df` 등으로 원래의 df를 copy()해서 진행할 것 - immutable

```python
# check NaN by row
merged_df.isnull().sum(axis=1) > 0
# Nan이 아닌 부분만 DataFrame으로 다시 생성
merged_df[merged_df.isnull().sum(axis=1) > 0]
nan_filled_merged_df = merged_df.copy()[~(merged_df.isnull().sum(axis=1) > 0)]
```

##### Scaling 
* 수치형 데이터인 열은 대부분 normalizaion 해주는 것이 좋다
* Scaling이 필요하다고 생각하는 수치형 데이터인 열을 찾아서 열별로 Standard Scaling 해준다
    * Statndard : x-열의 평균/열의 표준편차
* 위의 과정을 시행할 때 scaled_df 등으로 원래의 df를 copy()해서 진행할 것 - immutable

```python
# Choose numeric columns
scaling_needed_cols = ['Total Volume', 'Value']
scaling_needed_df = nan_filled_merged_df[scaling_needed_cols]

# 위의 컬럼들을 Scaling
scaling_needed_df = (scaling_needed_df - scaling_needed_df.mean()) / scaling_needed_df.std()

# scaled_df 에 반영
scaled_df = nan_filled_merged_df.copy()
scaled_df[scaling_needed_cols] = scaling_needed_df
```

##### Save DataFrame

```python
with open('BathSoapPreprocessed.df', 'wb') as f:
    pickle.dump(scaled_df, f)
```

### 3. Basic Visualization : Scatter Plot of Normalized Total Volume vs. Normalized Value
---
#### 3-1. `matplotlib.pyplot`

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(scaled_df['Value'], scaled_df['Total Volume'] )
plt.xlabel('Value', )
plt.ylabel('Total Volume')
plt.show()
```

#### 3-2. `bokeh.js`

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import HoverTool, ColumnDataSource

title = 'Total Volume vs. Value '
xaxis_label = 'Total Volume'
yaxis_label = 'Value'
x_range = [-2, 5]
y_range = [-2, 5]


output_notebook()

source = ColumnDataSource(data = dict (
    x = scaled_df[xaxis_label],
    y = scaled_df[yaxis_label],
    member_id = scaled_df['MEM']
))

# figure
TOOLS="hover,crosshair,pan,wheel_zoom,box_zoom,reset,tap,previewsave,box_select"
p = figure(title=title, tools=TOOLS, x_range = x_range, y_range = y_range, x_axis_type=None)

# plot kind
p.scatter('x', 'y', source=source)


p.xaxis.axis_label = 'Normalized' + xaxis_label
p.yaxis.axis_label = 'Normalized' + yaxis_label

# hover Setting
hover = p.select_one(dict(type=HoverTool))

hover.point_policy = "follow_mouse"
hover.tooltips = [
    (xaxis_label, "@x"),
    (yaxis_label, "@y"),
    ('Member Id', "@member_id")
]

show(p)
```
