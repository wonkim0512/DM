### 1. Data
#### 1-1.  `german_predictions.csv`

`german_predictions.csv`는 독일 은행에 신용거래를 신청한 신청자 1000명의 데이터이다. 관찰번호(OBS#)를 제외하고 총 28개의 변수로 이루어져있으며, 각 신청자들의 '신용 금액(AMOUNT)'가 예측하고자 하는 변수이다.

```python
import pandas as pd
```

```python
# 변수별 설명 : CodeList
pd.read_csv("CodeList_predictions.csv").iloc[:, :4]
```

15번 dummify면서 여성은 안 나타남.

우리가 prediction하려는 목표가 30번 amount

```python
# df
df = pd.read_csv("german_predictions.csv").set_index("OBS#")
df.tail()
```

```python
list(df.columns)
```

#### 1-2. Preprocessing : dummy variables

```python
# dummy varaibles : preprocessed_df
want_to_dummify_cols = ["CHK_ACCT", "HISTORY", "SAV_ACCT", 'EMPLOYMENT', 'JOB'] # 해당 범주형 변수들은 순서가 있긴 하지만 그 평균 등이 의미를 갖지는 않으므로 더미화하자.
preprocessed_df = pd.get_dummies(df,columns = want_to_dummify_cols)
preprocessed_df
```

```python
# split X, y -> prediction에서는 우리가 맞추고 싶은 답이 있다! 바로 amount가 y가 되겠지.
X, y = preprocessed_df.drop("AMOUNT", axis = 1), preprocessed_df.AMOUNT
```

#### 1-3. Train-Test Split : `sklearn.model_selection.train_test_split`
---
지도학습을 위해 주어진 데이터를 train set과 test set으로 나눈다.

```python
# import packages
from sklearn.model_selection import train_test_split
```

```python
# train_test_split  # index를 기반으로 쪼개줌.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 23)
```

random_state: 여러가지 random한 state가 있는데 그중에 42번째 형태를 사용하겠다. 그래서 동일한 수행결과를 보장함으로써 연구에 편함. 실험할 때마다 다른 모습을 보이면 골치아프겠지.

### 2. Linear Regression
---

#### 2-1. `sklearn.linear_model.LinearRegression`
---

```python
# import packages
from sklearn.linear_model import LinearRegression
```

```python
# Parameters
# fit_intercept = True # 상수를 넣을 것인지
# normalize = False # 변수들을 normalize한 상태로 모델을 얻을 것인지
```

```python
# lr_model
lr_model = LinearRegression()
```

```python
# fit to model
lr_model.fit(X = X_train, y = y_train)
```

```python
# get predictions
y_pred_lr = lr_model.predict(X = X_test)
```

```python
lr_model.coef_ # 가변수화를 시켰으므로 갯수가 늘어남
lr_model.intercept_ # 절편은 이렇게
```

```python
# result_y_test_df 
result_y_test_df = pd.DataFrame(y_test)
result_y_test_df['pred_lr'] = y_pred_lr
result_y_test_df
```

#### 2-2. Evaluation Measure 'RMS' : `sklearn.metrics.mean_squared_error`
---

```python
# import packages
from sklearn.metrics import mean_squared_error
import math
```

```python
# error_lr
error_lr = mean_squared_error(y_pred = y_pred_lr, y_true = y_test)
error_lr
```

```python
# rms_lr
rms_lr = math.sqrt(error_lr)
rms_lr
```

#### 5-2.  K-Fold Cross Validation : `sklearn.cross_validation.cross_val_score`
---
LR, KNN, SVM 등 모델 등 중에 어떤 모델을 사용할지 결정할 때 사용한다. 지금까지는 한가지 경우에 대해서만 모델을 평가했기 때문에.

Data가 적을 때 CV는 유용하게 사용됨. 결국 data가 귀하기 때문에!

```python
# import packages
from sklearn.cross_validation import cross_val_score
```

```python
# set parameters
estimator= lr_model # fit할 모델
cv= 5               # 전체 데이터를 k개의 그룹으로 나누고 총 k번의 모델 training - test를 반복한다.
                    # 이 때 각 회차마다 test set이 되는 그룹을 계속 바꾼다.

    
def rmse_scorer(estimator, X, y):
    y_pred = estimator.predict(X = X)
    rmse = math.sqrt(mean_squared_error(y_pred = y_pred, y_true = y))
    return rmse

scorer= rmse_scorer            # Evaluation Measure로 사용할 Score Function - scorer(estimator, X, y). 이걸 만들어줘야해
```

```python
# cross_val_score
cross_val_score(estimator = estimator, cv = cv, scoring = rmse_scorer, X = X, y = y) #cv는 X를 통째로 던져주면 되겠지!
```

```python
# cv_test_result_df
cv_test_result_df = pd.DataFrame()
cv_test_result_df[estimator] = cross_val_score(estimator = estimator, cv = cv, scoring = rmse_scorer, X = X, y = y)
cv_test_result_df
```

```python
cv_test_result_df.mean()
```
